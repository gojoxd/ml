# Checking VIF values

```python
def calculate_vif(x_var):
    for i in range(x_var.shape[1]):
        x = x_var.iloc[:, i]
        y = x_var.drop(x_var.columns[i], axis=1)
        r_squared = sm.OLS(x, y).fit().rsquared
        vif = round(1 / (1 - r_squared), 2)
        print(x_var.columns[i], 'VIF =', vif)

# Example usage:
calculate_vif(df1)
```
