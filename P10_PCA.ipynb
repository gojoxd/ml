```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn import datasets
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris=datasets.load_iris()
iris
```

```python
x=iris.data
```

```python
scaler=StandardScaler()
```

```python
ScaledData=pd.DataFrame(scaler.fit_transform(x))
```

```python
ScaledData
```

```python
sns.heatmap(ScaledData.corr())
```

```python
pca=PCA(n_components=3)
```

```python
pcadata=pca.fit(ScaledData)
```

```python
pcadata=pca.transform(ScaledData)
Dim_reduced_data=pd.DataFrame(pcadata)
Dim_reduced_data
```

```python
sns.heatmap(Dim_reduced_data.corr())
```
